```python
df_scan_splited_list_by_label = []
for i in range(10):
    df = df_scan[df_scan['label_2']== i]
    df_scan_splited_list_by_label.append(df)
    
merge_list = []
    
for table in df_scan_splited_list_by_label:
    train, validate, test = np.split(table.sample(frac=1), [int(len(table)*14/20), int(len(table)*17/20)])
    train['set_num'] = str(0)
    validate['set_num'] = str(1)
    test['set_num'] = str(2)
    splited_tabel = pd.concat([train, validate, test])
    merge_list.append(splited_tabel)
    
df_scan_splited_set = pd.concat(merge_list)
```
