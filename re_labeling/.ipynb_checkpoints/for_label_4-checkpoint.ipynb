```python
import pandas as pd
import numpy as np
import pickle
np.random.seed(4444)  
```

```python
df = pd.read_csv('../../../../jaeyeun/01_nh_poc/16_re_label/split_70_15_15_re_2.csv')
```

```python
df
```

* label copy (3->4)

```python
df['label_4'] = df['label_3']
```

```python
df
```

* sort column sequence

```python
df = df[['file_name', 'raw_text', 'prep_v9_text', 'dataset_n', 'train_val_test', 'label_1', 'label_2', 'label_3', 'label_4']]
```

```python
df
```

##  label_4

* 1st, assign label_1 to legal doc.

```python
legal = df[df['dataset_n'] != 0]
```

```python
for i, data in legal.iterrows():
    df.loc[i, 'label_4'] = data['label_1']
```

```python
df
```

* 2nd, assing label = 9 to data which contain '사회통계국 인구동향과'

```python
l = []
for i, row in df.iterrows():
    if row['label_3'] == 6 or row['label_3'] == 9:
        # if '인구' in row['raw_text'] and '동향' in row['raw_text']:
        if '담당부서 사회통계국 인구동향과' in row['raw_text']:
            l.append(row['file_name'])
len(l)
```

```python
for file_name in l:
    df.loc[df[df['file_name'] == file_name].index[0], 'label_4'] = 9
```

```python
#결과 체크
l = []
for i, row in df.iterrows():
    if row['label_4'] == 6:
        # if '인구' in row['raw_text'] and '동향' in row['raw_text']:
        if '담당부서 사회통계국 인구동향과' in row['raw_text']:
            l.append(row['file_name'])
len(l)
```

* 여기부터 re split 

```python
df
```

* remove original 'train_val_test' column

```python
df_wo_tvt = df.drop(columns=['train_val_test'], inplace=False)
```

```python
df_wo_tvt.head()
```

* seperate into scan, kdi 

```python
df_scan = df_wo_tvt[df_wo_tvt['dataset_n'] != 0]
```

```python
df_scan.head()
```

```python
df_kdi = df_wo_tvt[df_wo_tvt['dataset_n'] == 0]
```

```python
df_kdi.head()
```

```python
df_scan_splited_list_by_label = []
for i in range(10):
    df = df_scan[df_scan['label_4']== i]
    df_scan_splited_list_by_label.append(df)
    
merge_list = []
    
for table in df_scan_splited_list_by_label:
    train, validate, test = np.split(table.sample(frac=1), [int(len(table)*14/20), int(len(table)*17/20)])
    train['train_val_test'] = str(0)
    validate['train_val_test'] = str(1)
    test['train_val_test'] = str(2)
    splited_tabel = pd.concat([train, validate, test])
    merge_list.append(splited_tabel)
    
df_scan_splited_set = pd.concat(merge_list)
```

```python
df_scan_splited_set
```

```python
print(len(df_scan_splited_set[df_scan_splited_set['label_4']==1][df_scan_splited_set['train_val_test']=='0']))
print(len(df_scan_splited_set[df_scan_splited_set['label_4']==1][df_scan_splited_set['train_val_test']=='1']))
print(len(df_scan_splited_set[df_scan_splited_set['label_4']==1][df_scan_splited_set['train_val_test']=='2']))
```

```python
df_kdi_splited_list_by_label = []
for i in range(10):
    df = df_kdi[df_kdi['label_4']== i]
    df_kdi_splited_list_by_label.append(df)
    
merge_list = []
    
for table in df_kdi_splited_list_by_label:
    train, validate, test = np.split(table.sample(frac=1), [int(len(table)*14/20), int(len(table)*17/20)])
    train['train_val_test'] = str(0)
    validate['train_val_test'] = str(1)
    test['train_val_test'] = str(2)
    splited_tabel = pd.concat([train, validate, test])
    merge_list.append(splited_tabel)
    
df_kdi_splited_set = pd.concat(merge_list)
```

```python
print(len(df_kdi_splited_set[(df_kdi_splited_set['label_4'] == 2) & (df_kdi_splited_set['train_val_test'] == '0')]))
print(len(df_kdi_splited_set[(df_kdi_splited_set['label_4'] == 2) & (df_kdi_splited_set['train_val_test'] == '1')]))
print(len(df_kdi_splited_set[(df_kdi_splited_set['label_4'] == 2) & (df_kdi_splited_set['train_val_test'] == '2')]))
```

```python
df_final = pd.concat([df_kdi_splited_set, df_scan_splited_set])
```
