```python
import pickle
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

* need to modify

```python
file_path = '../../../../jaeyeun/01_nh_poc/60_BERT_NH_NO_3/nh_output_20200902_prep_v10_relabel_3_shuffled_epoch_7/test_set_prediction.pickle'
save_path = './conf_mat_exp3_prep10_label4_BERT_only-kdi.xlsx'
```

* seperate kdi and legal

```python
with open(file_path, 'rb') as f:
    pred_dic = pickle.load(f)
```

```python
y_true = []
y_pred = []

for key in pred_dic.keys():
    if pred_dic[key]['dataset_n'] != '0':
        continue
    y_true.append(pred_dic[key]['label'])
    y_pred.append(pred_dic[key]['pred'])
```

```python
y_true = pd.Series(y_true, name="True")
y_pred = pd.Series(y_pred, name="Pred")
df_confusion = pd.crosstab(y_true, y_pred)
```

```python
metric_dic = classification_report(y_true, y_pred, output_dict=True)
```

```python
precision = []
recall = []
f1_score = []
support = []

for key in metric_dic.keys():
    if key == 'accuracy':
        continue
    precision.append(metric_dic[key]['precision'])
    recall.append(metric_dic[key]['recall'])
    f1_score.append(metric_dic[key]['f1-score'])
    support.append(metric_dic[key]['support'])
```

```python
df_confusion['precision'] = precision[:len(df_confusion)]
df_confusion['recall'] = recall[:len(df_confusion)]
df_confusion['f1_score'] = f1_score[:len(df_confusion)]
df_confusion['support'] = support[:len(df_confusion)]
```

```python
df_confusion.loc['macro_avg', 'precision'] = metric_dic['macro avg']['precision']
df_confusion.loc['macro_avg', 'recall'] = metric_dic['macro avg']['recall']
df_confusion.loc['macro_avg', 'f1_score'] = metric_dic['macro avg']['f1-score']
df_confusion.loc['weighted_avg', 'precision'] = metric_dic['weighted avg']['precision']
df_confusion.loc['weighted_avg', 'recall'] = metric_dic['weighted avg']['recall']
df_confusion.loc['weighted_avg', 'f1_score'] = metric_dic['weighted avg']['f1-score']
```

```python
df_confusion.loc['accuracy', 'accuracy'] = metric_dic['accuracy']
```

```python
df_confusion.to_excel(save_path, index=True)
```
